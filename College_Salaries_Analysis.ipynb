```python
response1 = requests.get('https://www.payscale.com/college-salary-report/majors-that-pay-you-back/bachelors')
soup1 = bs(response1.text, 'html.parser')

Undergrad_major_scrape_raw1 = soup1.find_all('span', class_='data-table__value')
Undergrad_major_scrape1 = []
for s in range(150):
    Undergrad_major_scrape1.append(Undergrad_major_scrape_raw1[s].getText().replace('$', '').replace(',', ''))

splits = np.array_split(Undergrad_major_scrape1, 25)
Undergrad_scrape_splitted1 = []
for array in splits:
    Undergrad_scrape_splitted1.append((list(array)))
    

Undergrad_Major_Final1 = []
Undergrad_Early_Career_Pay1 = []
Undergrad_Mid_Career_Pay1 = []

for x in range(25):
    for y in range(1):
        Undergrad_Major_Final1.append(Undergrad_scrape_splitted1[x][1])
        Undergrad_Early_Career_Pay1.append(float(Undergrad_scrape_splitted1[x][3]))
        Undergrad_Mid_Career_Pay1.append(float(Undergrad_scrape_splitted1[x][4]))

# print(Undergrad_major_scrape1)
# print(Undergrad_scrape_splitted1)
print(Undergrad_Major_Final1)
print(Undergrad_Early_Career_Pay1)
print(Undergrad_Mid_Career_Pay1)
```

```python
response2 = requests.get('https://www.payscale.com/college-salary-report/majors-that-pay-you-back/bachelors/page/2')
soup2 = bs(response2.text, 'html.parser')

Undergrad_major_scrape_raw2 = soup2.find_all('span', class_='data-table__value')
Undergrad_major_scrape2 = []
for s in range(150):
    Undergrad_major_scrape2.append(Undergrad_major_scrape_raw2[s].getText().replace('$', '').replace(',',''))

splits = np.array_split(Undergrad_major_scrape2, 25)
Undergrad_scrape_splitted2 = []
for array in splits:
    Undergrad_scrape_splitted2.append((list(array)))
    

Undergrad_Major_Final2 = []
Undergrad_Early_Career_Pay2 = []
Undergrad_Mid_Career_Pay2 = []

for x in range(25):
    for y in range(1):
        Undergrad_Major_Final2.append(Undergrad_scrape_splitted2[x][1])
        Undergrad_Early_Career_Pay2.append(float(Undergrad_scrape_splitted2[x][3]))
        Undergrad_Mid_Career_Pay2.append(float(Undergrad_scrape_splitted2[x][4]))


# print(Undergrad_major_scrape2)
# print(Undergrad_scrape_splitted2)
print(Undergrad_Major_Final2)
print(Undergrad_Early_Career_Pay2)
print(Undergrad_Mid_Career_Pay2)
```

```python
Undergrad_Major_Final = Undergrad_Major_Final1 + Undergrad_Major_Final2
Undergrad_Early_Career_Pay = Undergrad_Early_Career_Pay1 + Undergrad_Early_Career_Pay2
Undergrad_Mid_Career_Pay = Undergrad_Mid_Career_Pay1 + Undergrad_Mid_Career_Pay2
```

```python
new_df_undergrad = pd.DataFrame(list(zip(Undergrad_Major_Final, Undergrad_Early_Career_Pay, 
                                         Undergrad_Mid_Career_Pay)), columns = 
                                ['Undergraduate Major', 'Early Career Pay', 'Mid-Career Pay'])
new_df_undergrad.head()
```

```python
high_potential_new = new_df_undergrad.sort_values('Mid-Career Pay',ascending=False)
high_potential_new[['Undergraduate Major', 'Mid-Career Pay']].head()
```
